```python
import os

import keras
from keras.layers import *
from keras.models import *
import mido
import numpy
import pandas
```

```python
numpy.set_printoptions(threshold=numpy.nan)
```

```python
NOTES = 36
```

```python
data_dir = '/home/santiago/Projects/ProjectEuterpe/data/chord_classification/'
midi_file = os.path.join(data_dir, 'Chords.mid')
csv_file = os.path.join(data_dir, 'Chords.csv')
```

```python
chord_types = pandas.read_csv(csv_file)
```

```python
chord_types
```

```python
chords = []
midi = mido.MidiFile(midi_file)
for track in midi.tracks:
    if track.name == 'Piano':
        new_chord = False
        next_chord = []
        for message in track:
            if message.type == 'note_on':
                if new_chord:
                    new_chord = False
                    chords.append(next_chord)
                    next_chord = []
                next_chord.append(message.note)
            if message.type == 'note_off':
                new_chord = True
        chords.append(next_chord)
```

```python
chords
```

```python
len(chords)
```

```python
chord_features = pandas.DataFrame(columns=['Name'])
```

```python
input_data = []
output_data = []
for i, chord in enumerate(chords):
    transpositions = NOTES - (max(chord) - min(chord))
    for j in range(transpositions):
        next_input = numpy.zeros((NOTES,), dtype=int)
        for note in chord:
            next_input[note - min(chord) + j] = 1
        next_output = numpy.zeros((len(chords),), dtype=int)
        next_output[i] = 1
        input_data.append(next_input)
        output_data.append(next_output)
        chord_features = chord_features.append({'Name': chord_types['Name'][i]}, ignore_index=True)
input_data = numpy.array(input_data, dtype=int)
output_data = numpy.array(output_data, dtype=int)
```

```python
chord_features
```

```python
assert len(input_data) == len(output_data)
```

```python
len(input_data)
```

```python
numpy.random.seed(0)
```

```python
input_layer = Input((NOTES,))
x = Dense(len(chords) * 4, activation='relu')(input_layer)
x = Dropout(0.5)(x)
x = Dense(len(chords) * 3, activation='relu')(x)
x = Dropout(0.5)(x)
x = Dense(len(chords) * 2, activation='relu')(x)
x = Dropout(0.5)(x)
output_layer = Dense(len(chords), activation='softmax')(x)
model = Model(input_layer, output_layer)
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
model.summary()
```

```python
model.fit(input_data, output_data, batch_size=1094, epochs=10000, shuffle=True)
```

```python
model.save_weights(os.path.join(data_dir, 'checkpoint.hdf5'))
```

```python
predictions = numpy.round(model.predict(input_data)).astype(int)
```

```python
indices = numpy.array(list(map(lambda x: numpy.argmax(x), predictions)), dtype=int)
```

```python
indices
```

```python
input_layer2 = Input((NOTES,))
x2 = Dense(len(chords) * 4, activation='relu', weights=model.layers[1].get_weights())(input_layer2)
x2 = Dropout(0.5)(x2)
x2 = Dense(len(chords) * 3, activation='relu', weights=model.layers[3].get_weights())(x2)
x2 = Dropout(0.5)(x2)
x2 = Dense(len(chords) * 2, activation='relu', weights=model.layers[5].get_weights())(x2)
output_layer2 = Dropout(0.5)(x2)
truncated = Model(input_layer2, output_layer2)
truncated.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
truncated.summary()
```

```python
features = truncated.predict(input_data)
```

```python
features.shape
```

```python
chord_features = pandas.concat([chord_features, pandas.DataFrame(features)], axis=1)
```

```python
chord_features
```

```python
chord_features.to_csv(os.path.join(data_dir, 'all_features.txt'), sep='\t', index=False)
```
